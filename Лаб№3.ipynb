**Задание 1.** Извлечение и преобразование данных (ETL) Необходимо преобразовать неструктурированную строку RAW_LOG_DATA в список словарей для дальнейшей обработки.

**Требования к реализации:**

1. Разделить исходную строку на отдельные записи.

2. Исключить строку заголовка (id,ip...).

3. Используя цикл, обработать каждую строку:

4. Удалить пробельные символы по краям.

5. Разделить строку по символу-разделителю (,).

6. Сформировать словарь для каждой записи с ключами: 'ip', 'time', 'user', 'status'.

**Задание 2.** Анализ сбоев аутентификации Реализовать алгоритм выявления подозрительной активности. Критерием атаки считать наличие более 2-х неудачных попыток (`fail`) входа с одного IP-адреса.

**Требования к реализации:**

1. Создать словарь failed_attempts для подсчета ошибок по IP-адресам.

2. Пройти циклом по списку parsed_logs.

3. Если статус операции — 'fail', увеличить счетчик для соответствующего IP.

4. Определить IP-адрес источника атаки (адрес, превысивший порог ошибок).

5. Сохранить его в переменную suspect_ip.

```python
parsed_logs = [
    {'ip': '192.168.1.5', 'time': '2024-05-20 10:00:01', 'user': 'admin', 'status': 'success'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:00:05', 'user': 'root', 'status': 'fail'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:00:06', 'user': 'root', 'status': 'fail'},
{'ip': '192.168.1.10', 'time': '2024-05-20 10:00:07', 'user': 'user1', 'status': 'success'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:00:08', 'user': 'root', 'status': 'fail'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:00:09', 'user': 'root', 'status': 'fail'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:00:10', 'user': 'root', 'status': 'fail'},
{'ip': '88.10.55.12', 'time': '2024-05-20 10:01:00', 'user': 'guest', 'status': 'fail'},
{'ip': '192.168.1.5', 'time': '2024-05-20 10:01:05', 'user': 'admin', 'status': 'success'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:02:00', 'user': 'root', 'status': 'success'},
{'ip': '88.10.55.12', 'time': '2024-05-20 10:02:05', 'user': 'guest', 'status': 'fail'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:03:00', 'user': 'admin', 'status': 'fail'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:03:01', 'user': 'admin', 'status': 'fail'},
{'ip': '45.33.12.99', 'time': '2024-05-20 10:03:02', 'user': 'admin', 'status': 'success'},
{'ip': '192.168.1.20', 'time': '2024-05-20 10:04:00', 'user': 'user2', 'status': 'success'}
]

# 1. словарь
failed_attempts = {}

# 2. Проходим циклом по списку
for entry in parsed_logs:
    # 3. Если 'fail', увеличиваем счетчик
    if entry['status'] == 'fail':
        ip = entry['ip']
        failed_attempts[ip] = failed_attempts.get(ip, 0) + 1

# 4. Определяем IP-адрес источника атаки (превысивший порог в 2 ошибки)
suspect_ip = None
threshold = 2

for ip, count in failed_attempts.items():
    if count > threshold:
        suspect_ip = ip
        break

# Вывод
print("Статистика неудачных попыток по IP-адресам:")
for ip, count in failed_attempts.items():
    status = "ПОДОЗРИТЕЛЬНЫЙ" if count > threshold else "нормальный"
    print(f"IP: {ip}, неудачных попыток: {count} - {status}")

print(f"\nIP-адрес источника атаки: {suspect_ip}")
```

## Ответы на контрольные вопросы (текстовый формат)

### 1. Отличие методов split() и strip() при обработке строк

**Метод split()** предназначен для разделения строки на составные части. Он анализирует строку, находит указанный разделитель (по умолчанию пробел) и возвращает список подстрок, расположенных между этими разделителями. Например, при обработке CSV данных split(',') разделит строку по запятым на отдельные поля.

**Метод strip()** выполняет другую функцию — он удаляет все пробельные символы (пробелы, табуляции, символы новой строки) в начале и в конце строки. Это классическая операция очистки данных от "мусорных" символов, которые могли появиться при форматировании текста.

**Главное отличие** заключается в результате работы: split() превращает одну строку в список нескольких строк, а strip() возвращает очищенную строку. На практике эти методы часто используются вместе: сначала strip() удаляет лишние пробелы по краям, затем split() разбивает очищенную строку на части.

### 2. Проверка наличия ключа в словаре

Существует несколько способов безопасно проверить существование ключа в словаре:

**Оператор in** — самый простой и читаемый способ. Конструкция "if key in dictionary" возвращает True, если ключ существует, и False в противном случае.

**Метод get()** — наиболее рекомендуемый подход для большинства ситуаций. Этот метод пытается получить значение по ключу, но если ключ отсутствует, возвращает значение по умолчанию (None или указанное вами). Например, dictionary.get('key', 0) вернет 0, если ключ 'key' не найден.

**Метод setdefault()** — более специфичный инструмент, который не только проверяет наличие ключа, но и автоматически устанавливает значение по умолчанию, если ключ отсутствует.

**Блок try-except** — используется в ситуациях, когда отсутствие ключа считается исключительной ситуацией, но это менее эффективно для простых проверок.

В контексте задания наиболее уместно использование метода get(), так как он позволяет элегантно реализовать подсчет количества событий: failed_attempts[ip] = failed_attempts.get(ip, 0) + 1.

### 3. Преимущества списка словарей перед списком списков

Использование списка словарей для хранения структурированных данных имеет ряд существенных преимуществ:

**Читаемость кода** значительно повышается, так как обращение к данным происходит по понятным именам полей (entry['ip'], entry['user']), а не по индексам (entry[1], entry[2]), значение которых нужно держать в голове.

**Устойчивость к изменениям структуры** данных — если в будущем потребуется добавить новое поле (например, "port" или "protocol"), существующий код продолжит работать, так как он обращается к данным по ключам, а не полагается на фиксированный порядок полей.

**Самодокументированность** кода — глядя на словарь, сразу понятно, какие данные он содержит. Ключи выступают в роли описания полей данных.

**Гибкость выборки** — легко извлекать только нужные поля или комбинировать данные из разных записей, используя ключи для доступа.

### 4. Модификация алгоритма для отслеживания ошибок по пользователям

Для изменения алгоритма с отслеживания по IP на отслеживание по пользователям потребуется внести следующие коррективы:

**Изменение ключа словаря** — вместо IP-адреса в качестве ключа нужно использовать имя пользователя. Словарь failed_attempts_by_user будет хранить количество неудачных попыток для каждого пользователя.

**Модификация логики подсчета** — при обходе списка parsed_logs проверяется статус 'fail', и увеличивается счетчик для конкретного пользователя: failed_attempts_by_user[user] = failed_attempts_by_user.get(user, 0) + 1.

**Критерий отбора остается тем же** — более двух неудачных попыток, но теперь анализируются пользователи, превысившие этот порог.

**Дополнительные возможности анализа** — при таком подходе можно выявлять не только атаки на конкретные учетные записи, но и распределенные атаки, когда один пользователь атакуется с разных IP-адресов. Также можно анализировать временные паттерны: как часто происходят неудачные попытки, в какое время суток и т.д.

Такой подход особенно полезен для выявления целенаправленных атак на конкретные учетные записи, что дополняет анализ по IP, выявляющий массовые атаки с одного источника.
